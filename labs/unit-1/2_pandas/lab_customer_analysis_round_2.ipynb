```python
#remove missing values of rows
data=data.drop(columns=['state']) 
```

```python
## drop all rows with nan values
data=data.dropna()
data.isna().sum()
```

```python
# Datetime format - Extract the months from the dataset and store in a separate column 
# Then filter the data to show only the information for the first quarter 
# ie. January, February and March. Hint: If data from March does not exist
# consider only January and February
data['effective_to_date'].dtype
```

```python
data['effective_to_date'] = pd.to_datetime(data['effective_to_date'], errors='coerce')
```

```python
data['effective_to_date'].dtype
```

```python
data.head()
```

```python
# Put all the previously mentioned data transformations into a function.
def clean_data(df):
    df.columns=[m.lower().replace(' ', '_') for m in df.columns]
    df=df.drop(columns=['unnamed:_0', 'state'])
    df=df.dropna()
    df['effective_to_date']=pd.to_datetime(df['effective_to_date'], errors='coerce')
    return df
```

```python
# Use function
clean_data(data)
```
